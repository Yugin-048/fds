```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Student_performance_data _.csv")
```

```python
print("GPA Statistics:")
print(df["GPA"].agg(["mean", "median", "min", "max", "skew"]))
```

```python
plt.figure()
sns.histplot(df["GPA"], bins=20)
plt.title("GPA Distribution")
plt.show()

plt.figure()
sns.boxplot(y=df["GPA"])
plt.title("GPA Outliers")
plt.show()

plt.figure()
sns.countplot(x="GradeClass", data=df)
plt.title("GradeClass Distribution")
plt.show()
```

```python
plt.figure()
sns.boxplot(x="Age", y="GPA", data=df)
plt.title("Age vs GPA")
plt.show()

plt.figure()
sns.boxplot(x="Gender", y="GPA", data=df)
plt.title("Gender vs GPA")
plt.show()

plt.figure()
sns.boxplot(x="Ethnicity", y="GPA", data=df)
plt.title("Ethnicity vs GPA")
plt.show()
```

```python
plt.figure()
sns.boxplot(x="ParentalEducation", y="GPA", data=df)
plt.title("Parental Education vs GPA")
plt.show()

plt.figure()
sns.boxplot(x="ParentalSupport", y="GPA", data=df)
plt.title("Parental Support vs GPA")
plt.show()

plt.figure()
sns.boxplot(x="Tutoring", y="GPA", data=df)
plt.title("Tutoring vs GPA")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x="StudyTimeWeekly", y="GPA", data=df)
plt.title("Study Time vs GPA")
plt.show()

for col in ["Extracurricular", "Sports", "Music", "Volunteering"]:
    plt.figure()
    sns.boxplot(x=col, y="GPA", data=df)
    plt.title(f"{col} vs GPA")
    plt.show()
```

```python
plt.figure()
sns.scatterplot(x="Absences", y="GPA", data=df)
plt.title("Absences vs GPA")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x="StudyTimeWeekly", y="GPA", hue="Tutoring", data=df)
plt.title("Study Time × Tutoring")
plt.show()

plt.figure()
sns.boxplot(x="ParentalSupport", y="GPA", hue="ParentalEducation", data=df)
plt.title("Parental Support × Education")
plt.show()

plt.figure()
sns.scatterplot(x="StudyTimeWeekly", y="GPA", hue="Extracurricular", data=df)
plt.title("Extracurricular × Study Time")
plt.show()
```

```python
num_cols = ["Age", "StudyTimeWeekly", "Absences", "GPA"]
corr = df[num_cols].corr()

plt.figure()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()

print("Correlation with GPA:")
print(corr["GPA"].sort_values(ascending=False))
```

```python
high_gpa = df[df["GPA"] > 3.8]
low_gpa = df[df["GPA"] < 0.5]

print("High GPA students:", len(high_gpa))
print("Low GPA students:", len(low_gpa))
```

```python
df["StudyEfficiency"] = df["GPA"] / df["StudyTimeWeekly"]
df["ActivityScore"] = df["Sports"] + df["Music"] + df["Volunteering"]
TOTAL_SCHOOL_DAYS = 180
df["AttendanceRatio"] = df["Absences"] / TOTAL_SCHOOL_DAYS

print(df[["StudyEfficiency", "ActivityScore", "AttendanceRatio"]].head())
```
